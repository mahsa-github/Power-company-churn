Data Visualization

Objective: To understand the distribution of electricity consumption among clients.

Rationale: This helps identify patterns in consumption, such as whether most clients consume a similar amount of electricity or if there are significant outliers.

```python
plt.figure(figsize=(10, 6))
sns.histplot(client_df['cons_12m'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Electricity Consumption in 12 Months')
plt.xlabel('Consumption (kWh)')
plt.ylabel('Frequency')
plt.show()
```

2. Churn Rate by Number of Products (Bar Plot)

Objective: To compare churn rates across different numbers of products a customer has.
Rationale: This helps identify if customers with more products are less likely to churn.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='nb_prod_act', y='churn', data=client_df, palette='Greens')
plt.title('Churn Rate by Number of Products')
plt.xlabel('Number of Products')
plt.ylabel('Churn Rate')
plt.show()
```

3- Price Sensitivity Analysis (Scatter Plot with Regression Line)

Objective: To visualize the relationship between peak variable price and electricity consumption.

Rationale: This helps determine if higher peak prices correlate with higher or lower consumption.

```python
merged_df = pd.merge(price_df, client_df, on='id')


plt.figure(figsize=(10, 6))
sns.regplot(x=merged_df['price_peak_var'], y=merged_df['cons_12m'], scatter_kws={'color':'orange'}, line_kws={'color':'red'})
plt.title('Price Sensitivity Analysis')
plt.xlabel('Peak Variable Price')
plt.ylabel('Consumption (kWh)')
plt.show()
```

4. Correlation Heatmap
Objective: To identify correlations between different pricing variables and consumption.

Rationale: This helps in understanding which pricing factors are most strongly associated with consumption.

Color Scheme: Use a cool-to-warm gradient (blue to red) to easily distinguish positive and negative correlations.

```python
plt.figure(figsize=(12, 8))
correlation_matrix = merged_df[['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var', 'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix', 'cons_12m']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
